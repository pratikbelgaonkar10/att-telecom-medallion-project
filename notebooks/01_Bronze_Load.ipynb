```python


# ==========================================
# Read Customer File
# ==========================================

customer_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", False)
    .csv("/Workspace/Users/pratikbelgaonkar10@gmail.com/ATT Telecom Project/customer_master.csv")
)

customer_df.printSchema()
customer_df.show(10, truncate=False)

# ==========================================
# Read Payment File
# ==========================================

payment_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", False)
    .csv("/Workspace/Users/pratikbelgaonkar10@gmail.com/ATT Telecom Project/payment_transactions.csv")
)

#payment_df.printSchema()
payment_df.show(10)

# ==========================================
# Save Customer Bronze Table
# ==========================================

customer_df.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bronze_customer")

# ==========================================
# Save Payment Bronze Table
# ==========================================

payment_df.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("bronze_payment")

spark.sql("SHOW TABLES").show()

customer_df = spark.table("bronze_customer")

payment_df = spark.table("bronze_payment")

```
